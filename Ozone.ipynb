## Reading datasource and python setup
Using the pandas and matplotlib packages in python needs some setup.
I choose the "ggplot" style for the plots to ensure a comprehensive look and feel on all plots.
To enlarge the plots i chose a bigger figure size as defined in the default settings.
Later in the report I started to use the plotly library as my main plotting library.

```python
# DEPENDENCIES
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# LOADING DATA
CSV = pd.read_csv("Ozone.csv")
```

```python
# Scatterplot
def Scatterplot():
    plotMean = CSV.plot.scatter('Year', 'Mean', color='r')
    plotMax = CSV.plot.scatter('Year', 'Maximum', color='g', ax=plotMean, label="Max")
    plotMin = CSV.plot.scatter('Year', 'Minimum', color='b', ax=plotMean, label="Min")
    plotLine = CSV.plot.line('Year', 'Mean', color='r',ax=plotMean, label="Mean Line")
    plotMean.set_ylabel("Ozone depletion area in square kilometres")
    plt.show()
```

```python
# Lineplot
def Lineplot():
    plotYear = CSV.plot.line('Year', 'Mean')
    plotYear.set_ylabel("Ozone depletion area in square kilometres")
    plt.show()
```

##  Visualization basics, chart types

I then started to read about the different types of visualisations like bar plots, line plots and the other basic visualisation plots.
To learn more about the different plot types i went through my data source and looked for a use case for each plot in my data.

### Line chart
The line bar can be used to show a continously positive value on the x axis, for instance time <cite id="8stsg">(Yi & Sapountzis, 2021)</cite>, and show the respective y points in a coherent line.
I used this to show that the ozone depletion area increased dramatically over the years (1979 - 2000). A line bar will also show missing values (or outliers) like in the year 1995.


```python
Lineplot()
```

### Scatter plot
A scatter plot is best used to show a relation between two variables <cite id="8p83r">(Castan, 2019)</cite>. They can also be extended to show the relation between n to 1 variable (mean, max, min to the year variable).
It will also visualize outliers quite easily like the value in year 1995 and some of the following years.

```python
Scatterplot()
```
